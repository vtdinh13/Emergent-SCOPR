```python
import os
from PIL import Image
```

```python
image_list = [i for i in os.listdir("images")]

new_image_list = []
for i in image_list:
  new = "/content/images/" + i
  new_image_list.append(new)

new_image_list
```

```python
model = YOLO("/content/drive/MyDrive/SCOPR/train/weights/best.pt")
results = model(new_image_list)
```

```python
# Visualize the results


for image_path, result in zip(image_list, results):
    # Plot results image
    im_bgr = result.plot()  # BGR-order numpy array
    im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image
    # Save results image to disk
    outpath = "/content/drive/MyDrive/SCOPR/test/"
    im_rgb.save(outpath + f'results_{image_path}')


```
